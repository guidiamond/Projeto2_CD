### Import de Bibliotecas

```python
import pandas as pd
import numpy as np
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

### Fun√ß√µes do Classificador

```python

def limpeza_de_dados(treinamento, coluna):
    dados_limpos = []
    for e in treinamento[coluna]:
        limpeza = str(e)
        limpeza = limpeza.lower()
        limpeza = limpeza.replace('\n', '')
        limpeza = limpeza.replace('t.co', '')
        limpeza = limpeza.replace('#', '')
        limpeza = limpeza.replace(':', '')
        limpeza = limpeza.replace(',', '')
        limpeza = limpeza.replace('.', '')
        limpeza = limpeza.replace('/', '')
        limpeza = limpeza.replace('//', '')
        limpeza = limpeza.replace('...', '')
        limpeza = limpeza.replace(';', '')
        limpeza = limpeza.replace('\'', '')
        limpeza = limpeza.replace('rt', '')
        limpeza = limpeza.split()
        for palavra in limpeza:
            if '@' in palavra:
                limpeza.remove(palavra)
        dados_limpos.append(limpeza)
    
    return dados_limpos

def contagem_palavras(dados_limpos):
    contagem = {}
    for e in dados_limpos:
        for all in e:
            if all not in contagem:
                contagem[all] = 1
            else:
                contagem[all] += 1
    
    return contagem

def frequencia_das_palavras(contagem, prob_total):
    probabilidade_palavra = {}
    soma = sum(contagem.values())
    for k in contagem:
        probabilidade_palavra[k] = np.log((contagem[k] + 1) / (prob_total + soma))

    return probabilidade_palavra

def calculador_probabilidade(frases_tratadas, frequencia_palavra_positivo, frequencia_palavra_negativo, probabilidade_positivo_total, probabilidade_negativo_total, prob_total):
    resultado = []
    laplace_zero = np.log(1 / prob_total)
    for e in frases_tratadas:
        conta_positivo = []
        conta_negativo = []
        conta_zero = []
        for all in e:
            for (k), (k2) in zip(laplace_probabilidade_positivo, laplace_probabilidade_negativo): #loop nas frequencias positivas
                if all in k: #se a palavra estiver no dicionario de frequencias positivas
                    conta_positivo.append(frequencia_palavra_positivo[k]) #adicionar a lista conta_positivo a probabilidade da palavra k
                elif all in k2:
                    conta_negativo.append(frequencia_palavra_negativo[k2])
                elif all not in k and all not in k2:
                    conta_zero.append(laplace_zero)
        prob_positivo = np.prod(conta_positivo) * probabilidade_positivo_total * np.prod(conta_zero)
        prob_negativo = np.prod(conta_negativo) * probabilidade_negativo_total * np.prod(conta_zero)
        if prob_positivo > prob_negativo:
            resultado.append(1)
        else:
            resultado.append(0)
            
    return resultado
```
